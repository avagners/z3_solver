## 1. Задача

По тропинке вдоль кустов  
Шли 11 хвостов.  
Сосчитать я также смог,  
Что шагало 30 ног.  
Это вместе шли куда-то  
Петухи и поросята.  
А теперь вопрос таков:  
Сколько было петухов?  
И узнать я был бы рад,  
Сколько было поросят?  

```python
# наши переменные (количества животных)
rooster, piglet = Ints('rooster piglet')

solve(rooster >= 1,  # не менее одного петуха
      piglet >= 1,   # не менее одного поросенка
      rooster + piglet == 11,  # всего ровно 11 хвостов
      2 * rooster + 4 * piglet == 30)  # всего 30 ног
```

## 2. Доп. задачи

1) В клетку посадили кроликов и фазанов. У  
животных 35 голов и 100 ног. Сколько в  
клетке кроликов и сколько фазанов?  

```python
# наши переменные (количества животных)
rabbit, pheasant = Ints('rabbit pheasant')

solve(rabbit >= 1,  # не менее одного кролика
      pheasant >= 1,   # не менее одного фазана
      rabbit + pheasant == 35,  # всего ровно 35 голов
      4 * rabbit + 2 * pheasant == 100)  # всего 100 лап
```

2) Торговец продает орехи двух сортов:  
одни по 90 центов, другие по 60 центо
за 
килограм . Он хочет получить   50
килограммов смеси по 72 цента   за
килограмм. Сколько для э  того
потребуется орехов каждого с  рта? 

```python
first, second = Ints('first second')

solve(first >= 1,  # не менее одного кг первого сорта
      second >= 1, # не менее одного кг первого сорта
      0.9 * first + 0.6 * second == 50 * 0.72)
```

3) На детской площадке 8 двух и трехколесных велосипедов. Всего у
них 21 колесо. Сколько двух- и сколько трехколесных велосипедов
на площадкеда.

```python
two_wheeled, three_wheeled = Ints('two_wheeled three_wheeled')

solve(two_wheeled >= 1,  # не менее одного двухколесного велосипеда
      three_wheeled >= 1, # не менее одного трехколесного велосипеда
      3 * three_wheeled + 2 * two_wheeled == 21)
```
